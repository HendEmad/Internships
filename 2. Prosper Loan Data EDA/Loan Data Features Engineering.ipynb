# Data Cleaning

```python
# Check null values
data.isnull().sum().sort_values(ascending = False)
```

```python
# Find the percentage of missing values in each column of the data
percent_missing = data.isnull().sum() * 100 / len(data)
# Create a data frame included each column and percentage of missing values on it
missing_value_data = pd.DataFrame({'column_name': data.columns,
                                 'percent_missing': percent_missing})
```

```python
# Sorting the columns according to the missing values percentage
missing_value_data.sort_values('percent_missing', inplace=True)
```

```python
# Find columns have more than 50% of missing values
missing_value_data[missing_value_data['percent_missing'] > 50.0]
```

```python
# delete features with more than 50%  null values
data.drop(columns = ['ClosedDate', 'CreditGrade', 'ProsperPrincipalOutstanding', 'ProsperPrincipalBorrowed', 'TotalProsperLoans',
         'TotalProsperPaymentsBilled', 'OnTimeProsperPayments', 'ProsperPaymentsOneMonthPlusLate', 'ProsperPaymentsLessThanOneMonthLate',
         'ScorexChangeAtTimeOfListing', 'LoanFirstDefaultedCycleNumber', 'GroupKey'], inplace = True)
```

```python
# Dropping data that have unique values
data.drop(columns=["ListingKey","ListingNumber","LoanKey","LoanNumber","MemberKey","DateCreditPulled", "ProsperRating (Alpha)"],inplace= True )
```

```python
# Check the data size after droppig features that have unique values
data.shape
```

```python
# Find null values summation in each feature
x = data.isnull().sum()
for i in range(len(x)) :
    print( str(data.columns[i]) + "  : " + str(x[i]) )
```

```python
# Find number of each value in each feature, to decide how to hanle missing values on it
def valuesCounting(column_name):
    print(data[column_name].value_counts())
```

```python
columns = ['BorrowerAPR', 'EstimatedEffectiveYield', 'EstimatedLoss', 'EstimatedReturn',
           'ProsperRating (numeric)', 'ProsperScore', 'BorrowerState', 'Occupation',
           'EmploymentStatus', 'EmploymentStatusDuration', 'CreditScoreRangeLower', 
           'CreditScoreRangeUpper', 'FirstRecordedCreditLine', 'CurrentCreditLines',
           'OpenCreditLines', 'TotalCreditLinespast7years', 'InquiriesLast6Months',
           'TotalInquiries', 'CurrentDelinquencies', 'AmountDelinquent', 'DelinquenciesLast7Years',
           'PublicRecordsLast10Years', 'PublicRecordsLast12Months', 'RevolvingCreditBalance',
           'BankcardUtilization', 'AvailableBankcardCredit', 'TotalTrades', 'TradesNeverDelinquent (percentage)',
           'TradesOpenedLast6Months', 'DebtToIncomeRatio']
for x in columns:
  print(valuesCounting(x))
  print('------------------------------------------')
```

```python
# Replace some features' missing values with mean
mean_replacement = ['EmploymentStatusDuration', 'CurrentCreditLines', 'OpenCreditLines', 'TotalCreditLinespast7years',
                   'InquiriesLast6Months', 'CurrentDelinquencies', 'PublicRecordsLast10Years', 'RevolvingCreditBalance',
                   'BankcardUtilization', 'AvailableBankcardCredit', 'TotalTrades', 'TradesNeverDelinquent (percentage)',
                   'TradesOpenedLast6Months', 'DebtToIncomeRatio', 'BorrowerAPR',
                    'EstimatedEffectiveYield', 'EstimatedLoss', 'EstimatedReturn']
for x in mean_replacement:
    data[x].fillna(data[x].mean, inplace=True)
```

```python
# Replace some features' missing values with median
Median_replacement = ['ProsperRating (numeric)', 'ProsperScore', 'BorrowerState', 'Occupation', 'EmploymentStatus', 'CreditScoreRangeLower',
                     'CreditScoreRangeUpper', 'FirstRecordedCreditLine', 'AmountDelinquent', 'DelinquenciesLast7Years']
for x in Median_replacement:
    data[x].fillna(data[x].median, inplace=True)
```

```python
# Replace some features' missing values with mode
mode_replacement = ['TotalInquiries', 'PublicRecordsLast12Months']
for x in mode_replacement:
    data[x].fillna(data[x].mode, inplace=True)
```

```python
# Make sure that there are no missing values in the data
data.isna().sum().sum()
```

```python
# Converting ListingCreationDate to datetime
data['ListingCreationDate'] = pd.to_datetime(data['ListingCreationDate'])
data['ListingCreationDate'].head()
```

```python
# Converting LoanOriginationDate to datetime
data['LoanOriginationDate'] = pd.to_datetime(data['LoanOriginationDate'])
data['LoanOriginationDate'].head()
```

```python
# Checking if there are any loans before july 2009
(2006,2007,2008,2009) in data['ListingCreationDate'].dt.year
```

```python
# Replace Listing Category Number with String
def replace_cat_num(numeric):
    list_dict = {0:'Not Available', 1:'Debt Consolidation',2:'Home Improvement',
                 3:'Business',4:'Personal Loan',5:'Student Use',6:'Auto',7:'Other',
                 8:'Baby&Adoption',9:'Boat',10:'Cosmetic Procedure',
                 11:'Engagement Ring',12:'Green Loans',13:'Household Expenses',
                 14:'Large Purchases',15:'Medical/Dental',16:'Motorcycle',17:'RV',
                 18:'Taxes', 19:'Vacation', 20:'Wedding Loans'}
    return list_dict[numeric]
```

```python
# Adding Column to Dataframe for Category String
data['ListingCategory'] = data['ListingCategory (numeric)'].apply(replace_cat_num)
```

```python
# Dropping Numeric Column
data.drop(['ListingCategory (numeric)'], axis=1, inplace=True)
```

```python
data['ListingCategory'].value_counts()
```

```python
# Current, Completed and Cancelled ---> completed (1)
# Otherwise ---> Not completed (0)
data['Loan_Status'] = np.where((data['LoanStatus'] == 'Current')|(data['LoanStatus'] == 'Completed')|(data['LoanStatus'] == 'Cancelled'), 1, 0)
```

```python
# Check values of the new column
data['Loan_Status'].value_counts()
```
