### 1. Drop missing values
not recommended, bacause we may lose some useful information

```python
nf1_data.dropna()
```

All the data is removed because every row in the dataset has at least one missing value.

So, we need to remove only columns that have at least one missing value

```python
# We will specify the columns only by setting the `axis` parameter to `1`
columns_with_na_dropped = nf1_data.dropna(axis = 1)
columns_with_na_dropped.head()
```

```python
# How much data did we lose?
print("Columns in original dataset: %d \n" %nf1_data.shape[1])
print("Columns in na's dropped dataset: %d \n" %columns_with_na_dropped.shape[1])
```

We lost much data by applying this approach, so this is not recommended

```python
nf1_data.head()
```

### 2. Filling in missing values automatically
We use `fillna` method to sill NaN values in the dataset automatically. We pass an argument to it to specify the value we want to replace the NaN values with

```python
# Replace all NaN values with 0
nf1_data.fillna(0).head(6)
```

We also can replace all NA's values with the value that comes directly after it in the same column by passin the value `bfill` to the `method` argument of the `fillna` method

```python
# then replace all the remaining na's with 0
nf1_data.fillna(method = 'bfill', axis = 0).head(10).fillna(0)
```
