```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
iris_data = pd.read_csv("iris.csv", index_col = "Id")
iris_data.head()
```

### Histograms

```python
sns.histplot(iris_data['Petal Length (cm)'])
```

### Density plots
- Kernal denisty estimate(KDE) - Smoothed histogram
- `sns.kdeplot`

```python
sns.kdeplot(data = iris_data['Petal Length (cm)'], shade = True)
```

### two-dimensional (2D) KDE plot 
`jointplot`

```python
sns.jointplot(x = iris_data['Sepal Length (cm)'], y = iris_data['Sepal Width (cm)'], kind = 'kde')
```

- the curve at the top of the figure is a KDE plot for the data on the x-axis (in this case, iris_data['Petal Length (cm)'])
- the curve on the right of the figure is a KDE plot for the data on the y-axis (in this case, iris_data['Sepal Width (cm)']).

### Color-coded plots

```python
sns.histplot(data = iris_data, x = 'Petal Length (cm)', hue = 'Species')
```

```python
sns.kdeplot(data = iris_data, x = 'Petal Length (cm)', hue = 'Species', shade = True)
```

 according to this dataset, we might even be able to classify any iris plant as Iris setosa (as opposed to Iris versicolor or Iris virginica) just by looking at the petal length: if the petal length of an iris flower is less than 2 cm, it's most likely to be Iris setosa!
